```python
#to each row I want to apply string splitting thing to investors, then I want to break them into new columns and rows
def split_investors(investor_str):
    investor_dictionary = {}
    investor_list = list(investor_str.split("|"))
    #print(investor_list)
    for investor_ugly_str in investor_list:
        #print(investor_ugly_str)
        investor = investor_ugly_str.split(":")[0]
        #print(investor)
        #note: bug here, need to figure out which row, with prints method seems fine for most
        if len(investor_ugly_str.split(":")) == 1:
            funding_contribution = 0
        else:
            funding_contribution = investor_ugly_str.split(":")[1]
        #print(funding_contribution)
        investor_dictionary[investor] = funding_contribution
    return investor_dictionary

investor_series = df["Investors"].apply(split_investors)
#maybe I could make a new dataframe with investor name, contribution as columns
investor_df = pd.DataFrame()

```
